## 1. Maximum, minimum and average (mean) Close price

```python
# Creating a list of banks
list_of_banks = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

### Maximum Close price

```python
for bank in list_of_banks:
    print(f'The closing maximum price for {bank} is: ')
    print(banks_data[bank]['Close'].max())
    print('\n')
```

### Minimum Close price

```python
for bank in list_of_banks:
    print(f'The closing minimum price for {bank} is: ')
    print(banks_data[bank]['Close'].min())
    print('\n')
```

### Average closing price

```python
for bank in list_of_banks:
    print(f'The closing average price for {bank} is: ')
    print(banks_data[bank]['Close'].mean())
    print('\n')
```

## 2. The returns for each bank's stock

**Rt = (Pt / Pt-1) - 1**

Rt - return per some time period

Pt - price at a time t (in this case, some date)

Pt-1 - price of a previous data

```python
# Makign an empty dataframe for returns
returns = pd.DataFrame()
```

```python
for bank in list_of_banks:
    returns[bank+' Return'] = banks_data[bank]['Close'].pct_change()

returns.head()
```

### Creating a pairplot using seaborn of the return dataframe

- Which stock stands out? 
- Why?

#### Importing seaborn

```python
import seaborn as sns
```

```python
sns.pairplot(returns[1:])
```

Okay, from this plot we can see that there is something wrong with CityGroup stock.

If we research what happened to CityGroup, we can see that in November 2008, CityGroup was insolvent. So, they had a huge crush, that is why we see this anomaly.

## 4. Which stock is the riskiest over the entire time period?

A basic measurement of riskiness of the stock, is the standard deviation of the returns of the stock.

* If you have very large standard deviation of the returns, it means that your stock is going up-and-down a lot. That means it is unsteady - very risky.

* If you have small standard deviation of the returns, your stock is likely stable.

```python
# Standard deviation for each Bank
returns.std()
```

It looks like the riskiest overall is CityGroup stock, which probably have something to do with that crash of 2009.

## 5. Which stock is the riskiest for year 2015?

```python
returns.loc['2015-01-01':'2015-12-31'].std()
```

From the data, it can be inferred that the riskiest stock for 2015 is Morgan Stanley. The safest stock appears to be WFC.

## 6. Create a line plot showing Close price for each bank for the entire index of time.

#### Importing Plotly and Cufflinks

```python
import plotly
import cufflinks as cf
cf.go_offline()
cf.set_config_file(offline=False,world_readable=True)
```
